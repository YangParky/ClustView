```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = '2'  #代表只使用第 个gpu

import torch

from omegaconf import OmegaConf
from util import profile_model

from models.ClustViewF import ClustView_cls as ClustView_Naive
from models.ClustView import ClustView_cls as ClustView_Resnet

data = OmegaConf.load('./output/ModelNet40/ClustView/modelnet40.yaml')
train = OmegaConf.load('./output/ModelNet40/ClustView/train_cls.yaml')

args = OmegaConf.merge(data, train)

print('Done!')
```

```python
device = torch.device("cuda")

model1 = ClustView_Naive(args).to(device)
model2 = ClustView_Resnet(args).to(device)

print('Done!')
```

```python
data = torch.rand(2, 1024, 3).to(device)
input = [torch.randn_like(data)]

batch_size = data.shape[0]

flops1, macs1, params1 = profile_model(model1, input)
print(f'GFLOPs\tGMACs\tParams.(M)')
print(f'{flops1/(float(batch_size)*1e9): .2f}\t{macs1/(float(batch_size)*1e9): .2f}\t{params1/1e6: .3f}')

flops2, macs2, params2 = profile_model(model2, input)
print(f'GFLOPs\tGMACs\tParams.(M)')
print(f'{flops2/(float(batch_size)*1e9): .2f}\t{macs2/(float(batch_size)*1e9): .2f}\t{params2/1e6: .3f}')
```
